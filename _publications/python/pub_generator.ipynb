```python
import bibtexparser
import os
import re
from datetime import datetime
```

```python
def bibtex_to_markdown(bibtex_entries, output_dir="markdown_pubs"):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    parser = bibtexparser.bparser.BibTexParser(common_strings=True)
    bib_database = bibtexparser.loads('\n'.join(bibtex_entries), parser=parser)

    for entry in bib_database.entries:
        title = entry.get('title', 'No Title').replace('{', '').replace('}', '')
        authors = entry.get('author', 'Unknown Author')
        journal = entry.get('journal', entry.get('fjournal', 'Unknown Journal'))
        year = entry.get('year', '1900')
        volume = entry.get('volume', '')
        number = entry.get('number', '')
        pages = entry.get('pages', '')

        # Get first two words (excluding "the"), only letters
        words = [w for w in re.findall(r'\b[a-zA-Z]+\b', title) if w.lower() != "the"]
        filename_words = words[:2] if len(words) >= 2 else words
        filename_base = ''.join(filename_words).lower()

        date = f"{year}-01-01"
        citation = f"{authors}. ({year}). &quot;{title}.&quot; <i>{journal}</i>. {volume}({number})."
        md_content = f"""---
title: "{title}"
collection: publications
category: manuscripts
permalink: /publication/{date}-{filename_base}
excerpt: 'This paper is about {title}.'
date: {date}
venue: '{journal}'
paperurl: ''
citation: '{citation}'
---
"""
        filename = f"{date}-{filename_base}.md"
        with open(os.path.join(output_dir, filename), "w") as f:
            f.write(md_content)
```
